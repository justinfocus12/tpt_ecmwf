### 2. Reanalysis

Set simulation parameters for the reanalysis (RA) simulation. We will start the simulation partway through year 1957 (remember, the years are 400 days long, so this doesn't correspond to our world) and save out every 0.5 days. We will start with a 500-day "burn-in" period to let the system settle onto its attractor. 

```python
dt_samp = 0.5 # Time step to save out
duration_burnin = 500.0 # Run 500 days of burn-in
t_init_burnin = (1957 + 0.2)*fpd["year_length"] # Start the burnin partway through 1957, in our pretend model world
duration_ra = 10.0*fpd["year_length"] # Run 200 years of reanalysis (abbreviated RA)
```

```python
x0 = np.zeros((1,7))
x0[0,-1] = t_init_burnin # Time is the last dimension
t_burnin = np.arange(0, duration_burnin, dt_samp)
x_burnin = crom.integrate(x0,t_burnin)
```

Now simulate past burnin

```python
rng = default_rng(seed=125)
x0 = x_burnin[:,-1] # End of burn-in = beginning of reanalyis
# Add random perturbations
n_pert = 2
pert_x0 = 0.0001*rng.normal(size=(n_pert,crom.xdim))
pert_x0[:,-1] = 0
x0 = x0 + pert_x0

t_save = np.arange(0,duration_ra,dt_samp)
x = crom.integrate(x0,t_save)
```

```python
obs2plot = x[:,:,1]**2 + x[:,:,2]**2
```

```python
duration2plot = 2*fpd["year_length"]
tidx2plot = np.where(t_save <= duration2plot)[0]
fig,axes = plt.subplots(figsize=(20,5), ncols=2, sharey=True, gridspec_kw={"width_ratios": [3,1] })
fig.set_facecolor("white")
fig.subplots_adjust(top=0.95)
ax = axes[0]
ax.set_ylim(ylim)
ax.set_xlabel("Time [days]")
ax.set_ylabel("Wave-1 energy")
ylim = [obs2plot.min(), obs2plot.max()]
yrange = ylim[1] - ylim[0]
ylim[0] -= 0.02*yrange
ylim[1] += 0.01*yrange

handles = []
for i in range(n_pert):
    ax = axes[0]
    h, = ax.plot(t_save[tidx2plot], obs2plot[i,tidx2plot], label=f"Initial condition {i+1}")
    handles += [h]
    ax.legend(handles=handles, bbox_to_anchor=(0.2,1.0), ncol=2)
    ax = axes[1]
    _ = ax.hist(obs2plot[i,:], bins=50, histtype='step', fill=False, orientation='horizontal') 
    ax.set_title(f"{int(duration_ra)}-day histogram")
    fig.savefig(f"timeseries_{i}")
```

```python
def plot_flow_field(x):
    b = crom.q["b"]
    # Given a single snapshot, plot the flow field
    xy,yx = np.meshgrid(np.linspace(0,2*np.pi,100), np.linspace(0,np.pi*b,50))
    xyc = 0.25*(xy[1:,1:] + xy[1:,:-1] + xy[:-1,1:] + xy[:-1,:-1])
    yxc = 0.25*(yx[1:,1:] + yx[1:,:-1] + yx[:-1,1:] + yx[:-1,:-1])
    psi_prims = dict({
        "01": np.sqrt(2)*np.cos(yxc/b),
        "02": np.sqrt(2)*np.cos(2*yxc/b),
        "11": np.sqrt(2)*np.cos(xyc)*np.sin(yxc/b),
        "12": np.sqrt(2)*np.cos(xyc)*np.sin(2*yxc/b),
    })
    psi = (
        crom.q["b"]*x[0]*psi_prims["01"]
        + crom.q["b"]*x[3]*psi_prims["02"]
        + crom.q["b"]/np.sqrt(2)*x[1]*psi_prims["11"]
        + crom.q["b"]/np.sqrt(2)*x[4]*psi_prims["12"]
    )
    fig,ax = plt.subplots(figsize=(17,5))
    ax.set_ylabel(r"$y$")
    ax.set_title(r"Streamfunction")
    #im = ax.pcolormesh(xy,yx,psi,cmap=plt.cm.jet)
    #fig.colorbar(im,ax=ax,label=r"$\psi$")
    ax.contourf(xyc,yxc,psi,cmap=plt.cm.jet)
    ax.set_xlabel(r"$x$")
    return fig,ax
```

```python
tidx_range = np.arange(190,220)
for tidx in tidx_range:
    fig,ax = plot_flow_field(x[0,tidx,:-1])
    fig.set_facecolor("white")
    ax.set_title(f"Streamfunction on day {(x[0,tidx,-1]-x[0,tidx_range[0],-1]):0.1f}")
    fig.savefig(f"sfn{tidx}")
    plt.close(fig)
```
