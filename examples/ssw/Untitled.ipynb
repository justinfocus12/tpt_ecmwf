```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,2*np.pi,20)
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.tight_layout()
plt.savefig("/scratch/jf4241/ecmwf_data/results/tpt_quantities/2022-08-23/sinfig")
```
