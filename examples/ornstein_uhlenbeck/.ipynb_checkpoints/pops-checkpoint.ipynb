```python
# Create a time-dependent system (OU process) on a (xdim,) system
fpd = {"period": 0.2, "noise_amp": 0.01, "damping": 1.5, "dt_sim": 0.001} 
model = model_ou.OUModel(fpd)
```

```python
dt_save = 0.01
tmax = 100.0
t_save = np.linspace(0, tmax, int(tmax/dt_save+0.5))
x0 = 0.02*np.ones((1,3))
x = model.integrate(x0, t_save, seed=12345)
```

```python
fig,axes = plt.subplots(ncols=2,nrows=2,figsize=(12,12),sharex='col',sharey='row')
xlim = np.max(np.abs(x[:,:,0]))*np.array([-1,1])
ylim = np.max(np.abs(x[:,:,1]))*np.array([-1,1])
axes[1,0].plot(x[0,40:,0],x[0,40:,1], color='black')
axes[1,0].axhline(y=0,color='gray',alpha=0.5)
axes[1,0].axvline(x=0,color='gray',alpha=0.5)
axes[1,0].set_xlim(xlim)
axes[1,0].set_ylim(ylim)
_ = axes[0,0].hist(x[0,:,0],orientation='vertical',bins=100)
_ = axes[1,1].hist(x[0,:,1],orientation='horizontal',bins=100)
```
